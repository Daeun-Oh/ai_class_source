```python
import numpy as np
from glob import glob
from PIL import Image

files = [ (file, file.replace(".npy", "").replace("data/", ""))  for file in glob("data/*")]
files
```

```python
targets = []
for i, item in enumerate(files):
    _, target = item
    targets.append(target)

targets
```

```python
import matplotlib.pyplot as plt
train_input = None
train_target = None
for i, item in enumerate(files):
    file, target = item
    data = np.load(file)
    index = targets.index(target)
    print(index, target)
    _targets = np.repeat(index, len(data))

    data = data.reshape(-1, 28, 28)
    items = None
    for j in range(len(data)):
        item = np.array(Image.fromarray(data[j]).resize((224, 224))).reshape(1, 224, 224)
        
        if j == 0:
            items = item
            plt.imshow(item[0], cmap='gray_r')
            plt.axis('off')
            plt.show()
        else:
            items = np.concatenate((items, item), axis=0)

        print(":", target, ":",j + 1, ":", len(data))
    if i == 0:
        train_input = items
        train_target = _targets
    else:
        train_input = np.concatenate((train_input, items), axis=0)
        train_target = np.concatenate((train_target, _targets))

    print(train_input.shape)    


train_input = np.expand_dims(train_input, axis=3)
train_input = np.repeat(train_input, 3, axis=3)

train_input.shape
```

```python
print(train_input.shape, train_target.shape)
```

```python

#  ,   
np.save("train.npy", train_input)
np.save("target.npy", train_target)
```
